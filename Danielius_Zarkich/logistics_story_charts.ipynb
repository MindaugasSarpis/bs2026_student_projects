# Logistics story charts

Three views aligned with the **Smart Holiday** project:

1. **Route health map** (Folium) — corridor segments green vs red from `delivery_scheduler` legs + pop-ups.
2. **Holiday Gantt** (Plotly) — naive vs shifted “smart” truck bars with gray **holiday zones**.
3. **Friction heatmap** (Seaborn) — year-at-a-glance **logistics friction** (holidays + small December bump).

Requires: `folium`, `seaborn`, `matplotlib` (see `requirements.txt`).

```python
from datetime import date

from holiday_viz import (
    build_friction_calendar_heatmap,
    build_route_health_map,
    build_schedule_gantt_figure,
    friction_series_for_corridor,
)

SHIP = date(2026, 11, 9)
ORIGIN, DEST = "LT", "DE"
CORRIDOR = ["LT", "PL", "DE"]
```

## 1. Route health interactive map

Segments use the same **leg countries** as `ROUTES` in `delivery_scheduler.py`. Red = that leg has a holiday conflict for the chosen ship date; green = no conflict on that leg.

```python
route_map = build_route_health_map(ORIGIN, DEST, SHIP)
route_map
```

## 2. Holiday Gantt (naive vs smart)

Gray bands = at least one corridor country has a public holiday that day. **Smart** bars shift by the same offset as `suggest_safe_ship_date` vs the naive ship date.

```python
gantt_fig = build_schedule_gantt_figure(ORIGIN, DEST, SHIP, horizon_days=21)
gantt_fig.show()
```

## 3. Cost of celebration — friction heatmap

Each day’s score = number of corridor countries with a holiday **+** a small extra weight in December (peak-season narrative). Darker = more **logistics friction**.

```python
import matplotlib.pyplot as plt

fig_hm = build_friction_calendar_heatmap(2026, CORRIDOR)
plt.show()
```
